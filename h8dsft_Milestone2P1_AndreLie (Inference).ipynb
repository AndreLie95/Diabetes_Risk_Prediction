## Model Inference

#### Import Library

````python
import numpy as np
import pandas as pd
import joblib
````

### Load Model

````python
with open('pipeline_clf.pkl', 'rb') as file_1:
  clf = joblib.load(file_1)
````

### Get Data for Model Inference

````python
# membuat data random mengunakan rentang data masing-masing kolom dari dataset
rng = np.random.default_rng(seed=2895)
Age = rng.integers(1.0, 7.0, size=10)
Sex = rng.integers(0.0, 1.0, size=10)
HighChol = rng.integers(0.0, 1.0, size=10)
CholCheck = rng.integers(0.0, 1.0, size=10)
BMI = rng.integers(12.0, 98.0, size=10)
Smoker = rng.integers(0.0, 1.0, size=10)
HeartDiseaseorAttack = rng.integers(0.0, 1.0, size=10)
PhysActivity = rng.integers(0.0, 1.0, size=10)
Fruits = rng.integers(0.0, 1.0, size=10)
Veggies = rng.integers(0.0, 1.0, size=10)
HvyAlcoholConsump = rng.integers(0.0, 1.0, size=10)
GenHlth = rng.integers(1.0, 5.0, size=10)
MentHlth = rng.integers(0.0, 30.0, size=10)
PhysHlth = rng.integers(0.0, 30.0, size=10)
DiffWalk = rng.integers(0.0, 1.0, size=10)
Stroke = rng.integers(0.0, 1.0, size=10)
HighBP = rng.integers(0.0, 1.0, size=10)

data_inf = pd.DataFrame({'Age' : Age,
                         'Sex' : Sex,
                         'HighChol' : HighChol,
                         'CholCheck' : CholCheck,
                         'BMI' : BMI,
                         'Smoker' : Smoker,
                         'HeartDiseaseorAttack' : HeartDiseaseorAttack,
                         'PhysActivity' : PhysActivity,
                         'Fruits' : Fruits,
                         'Veggies' : Veggies,
                         'HvyAlcoholConsump' : HvyAlcoholConsump,
                         'GenHlth' : GenHlth,
                         'MentHlth' : MentHlth,
                         'PhysHlth' : PhysHlth,
                         'DiffWalk' : DiffWalk,
                         'Stroke' : Stroke,
                         'HighBP' : HighBP})

data_inf
````

### Inferencing

Prediction Result

````python
# menampilkan hasil prediksi dalam dataframe
prediction = clf.predict(data_inf)
data_inf['DiabetesRisk'] = prediction
data_inf
````
